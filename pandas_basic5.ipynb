##creating series object from a dic

```python
marks_1 = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks_1,name='shubhangi ke marks')
marks_series
```

```python
marks_2 = {
    'maths':67,
    'english':57,
    'science':np.nan,
    'hindi':100
}

marks_series_2 = pd.Series(marks_2,name='shubhangi ke marks')
marks_series_2
```
